```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_local = pd.read_csv(r'D:\My Hoa\work\VideoToAudioGuide\outputs\reports\test1_20250319_152519\execution_time_20250320_163533.csv')
df_api = pd.read_csv(r"D:\My Hoa\work\VideoToAudioGuide\outputs\reports\test1_20250317_170310\execution_time_20250320_164849.csv")
```

```python
def plot_average_comparison(df1, df2, label1="Dataset 1", label2="Dataset 2"):
    avg_times1 = df1.drop(columns=["frame_index"]).mean()
    avg_times2 = df2.drop(columns=["frame_index"]).mean()

    x = avg_times1.index
    width = 0.35  # Độ rộng cột

    plt.figure(figsize=(8, 5))
    bars1 = plt.bar(x, avg_times1, width=width, label=label1, color="blue")
    bars2 = plt.bar([i + width for i in range(len(x))], avg_times2, width=width, label=label2, color="orange")

    plt.xlabel("Task")
    plt.ylabel("Average Execution Time (ms)")
    plt.title("Average Execution Time Comparison")
    plt.xticks([i + width / 2 for i in range(len(x))], x, rotation=45)
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Ghi giá trị lên đầu mỗi cột
    for bars in [bars1, bars2]:
        for bar in bars:
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,
                     f"{bar.get_height():.2f}", ha="center", fontsize=10, fontweight="bold")

    plt.show()
```

```python
plot_average_comparison(df_local, df_api, "Local Host", "API")
```
